```python
import matplotlib.pyplot as plt

epsilons = [0, 0.02,0.04,0.06,0.08,0.1]
```

```python
plt.title("Epsilon vs Accuracy")
cnn_a_baseline = [0.80374205, 0.36942676, 0.19715366, 0.13783838, 0.10798169, 0.09375]
plt.plot(epsilons, cnn_a_baseline, "-",label="CNN (Baseline)")
cnn_a_002 = [0.703125, 0.7158638, 0.72979695, 0.7425358, 0.7494029, 0.7548766]
plt.plot(epsilons, cnn_a_002, "-",label="CNN (\u03B5=0.2)")
rnn_a_baseline = [0.79000795, 0.35459793, 0.21367437, 0.16709793, 0.14092356, 0.1267914]
plt.plot(epsilons, rnn_a_baseline, "-",label="CNN-LSTM (Baseline)")
rnn_a_002= [0.69685507, 0.7090963, 0.71984476, 0.73099124, 0.7376592, 0.7405454]
plt.plot(epsilons, rnn_a_002, "-",label="CNN-LSTM (\u03B5=0.2)")
plt.xlabel("Epsilon on Test Data")
plt.ylabel("Accuracy %")
plt.legend(loc="lower left", fontsize='small')
plt.show()
```

```python
cnn_a_diff = []
rnn_a_diff = []

for a_baseline, a_002 in zip(cnn_a_baseline, cnn_a_002):
    cnn_a_diff.append(a_002-a_baseline)

print(cnn_a_diff)

for a_baseline, a_002 in zip(rnn_a_baseline, rnn_a_002):
    rnn_a_diff.append(a_002-a_baseline)

print(rnn_a_diff)

plt.title("Epsilon vs Accuracy \u0394 for \u03B5=0.2 Trained Models")
plt.plot(epsilons, cnn_a_diff, "-",label="CNN")
plt.plot(epsilons, rnn_a_diff, "-",label="CNN-LSTM")
plt.xlabel("Epsilon on Test Data")
plt.ylabel("Accuracy \u0394 %")
plt.legend(loc="lower left", fontsize='small')
plt.show()
```
